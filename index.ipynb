# Harmonic Analysis

## Fourier trigonometric series

* Any function (or signal) $s(t)$ can be represented by a a linear combination of sinosuidal functions:

  <a id='Fourier_trigonometric_series'></a>
  \begin{equation}
    s(t) = a_0 + \sum_{i=1}^\infty \big(a_i \cos(i\omega_0t) + b_i
       \sin(i\omega_0t)\big), \tag{Fourier_trigonometric_series}
       \label{Fourier_trigonometric_series}
  \end{equation}

  where $$t_0<t<t_0+\frac{2\pi}{\omega_0},$$ $\omega_0$ if the *fundamental frequency* expresed in radians/second and $a_0$, $\{a_i\}$ and $\{b_i\}$ are the coefficients of the Fourier trigonometric series.

* Eq. [Fourier_trigonometric_series](#Fourier_trigonometric_series) only can reconstruct $s(t)$ in a interval of duration $\frac{2\pi}{\omega_0}$ which is equal to the period of the lowest frequency component of $s(t)$.

* The coefficients of the Fourier trigonometric series quantify the correlation between the signal $s(t)$ and the corresponding basis function:

  <a id='cosine_correlation'></a>
  \begin{equation}
    a_n=\frac{\displaystyle
      \int_{t_0}^{t_0+T}s(t)\cos(n\omega_0t)dt
      }{\displaystyle                                                             
      \int_{t_0}^{t_0+T}\cos^2(n\omega_0t)dt
      }
    \tag{cosine_correlation}
    \label{cosine_correlation}                                                               
  \end{equation}

  <a id='sine_correlation'></a>
  \begin{equation}
    b_n=\frac{\displaystyle
      \int_{t_0}^{t_0+T}s(t)\sin(n\omega_0t)dt
      }{\displaystyle                                                             
      \int_{t_0}^{t_0+T}\sin^2(n\omega_0t)dt
      }
    \tag{sine_correlation}
    \label{sine_correlation}                                                               
  \end{equation}

  where

  $$
    T=\frac{2\pi}{\omega_0}
  $$
  
  is the length of the interval where the analysis of $s(t)$ is carried out.

* Coefficient

  \begin{equation*}
    a_0 = \frac{1}{T}\int_{t_0}^{t_0+T} fst)dt                                 
  \end{equation*}

  is the average of $s(t)$ in $[t_0, t_0+T]$, and is also refeered as the DC (Direct Current) frequency coefficient of $s(t)$.

* Knowing that:

  $$                                                                            
    \int_{t_0}^{t_0+T} \cos^2(n\omega_0t)dt = \int_{t_0}^{t_0+T} \sin^2(n\omega_0t)dt = \frac{T}{2},                                           
  $$

  Eqs. [cosine_correlation](#cosine_correlation) and [sine_correlation](#sine_correlation) can be written as:

  <a id='cosine_correlation_Lathi'></a>
    \begin{equation}                                                              
      a_n = \frac{2}{T}\int_{t_0}^{t_0+T} s(t)\cos(n\omega_0t)dt                  
      \tag{cosine_correlation_Lathi}
      \label{cosine_correlation_Lathi}
    \end{equation}

  <a id='sine_correlation_Lathi'></a>
    \begin{equation}                                                              
      b_n = \frac{2}{T}\int_{t_0}^{t_0+T} s(t)\sin(n\omega_0t)dt,                 
      \tag{sine_correlation_Lathi}                                                                  
      \label{sine_correlation_Lathi}                                                             
    \end{equation}

  (see [[B.P. Lathi, 2005]](https://scholar.google.es/scholar?hl=es&as_sdt=0%2C5&q=lathi+linear+systems+and+signals&btnG=&oq=+lathi+li) and [[A.V. Oppenheim, 1999]](https://scholar.google.es/scholar?hl=es&as_sdt=0%2C5&q=oppenheim+discrete+time+signal+processing&btnG=&oq=Oppenheim+Discrete-time+sign)).

## Fourier exponential series

* Let the definitions:

  <a id='Fourier_series_relation'></a>
  \begin{equation}
    \begin{array}{l}
      a_0 = S_0\\
      a_n = S_{n}+S_{-n}\\
      b_n = j(S_{n}-S_{-n}),
    \end{array}
    \tag{Fourier_series_relation}
    \label{Fourier_series_relation}
  \end{equation}

  where $j=\sqrt{-1}$.

* Replacing Eqs. [Fourier_series_relation](#Fourier_series_relation) in Eq. [Fourier_trigonometric_series](#Fourier_trigonometric_series) we get that:

  \begin{equation}
    s(t) = S_0 + \sum_{n=1}^\infty \big((S_{n}+S_{-n}) \cos(n\omega_0t) + (j(S_{n}-S_{-n})) \sin(n\omega_0t)\big).                                     
  \end{equation}

* Multiplying:

  \begin{equation}
    \begin{array}{r}
      s(t) = S_0 + \displaystyle\sum_{n=1}^\infty \Big(S_{n}\cos(n\omega_0t)+S_{-n}\cos(n\omega_0t) +\\                                                          
      jS_{n}\sin(n\omega_0t)-jS_{-n}\sin(n\omega_0t)\Big).
    \end{array}                        
  \end{equation}

* Operating:

  \begin{equation}
    \begin{array}{r}
      s(t) = S_0 + \displaystyle\sum_{n=1}^\infty \Big(S_{n}\big(\cos(n\omega_0t) + j\sin(n\omega_0t)\big)+\\                                                      
      S_{-n}\big(\cos(n\omega_0t) -j\sin(n\omega_0t)\big)\Big).
  \end{array}                                                                   
  \end{equation}

* Applying in this equation the [trigonometric identities](https://en.wikipedia.org/wiki/List_of_trigonometric_identities#Relation_to_the_complex_exponential_function):

  <a id='exp_sin_cos_itentity'></a>
  \begin{equation}
    \begin{array}{c}
      e^{jn\omega_0t} = \cos(n\omega_0t) + j\sin(n\omega_0t)\\
      e^{-jn\omega_0t} = \cos(n\omega_0t) - j\sin(n\omega_0t)
    \end{array}
    \tag{exp_sin_cos_identity}                                                             
    \label{exp_sin_cos_identity}
  \end{equation}

  we obtain that:

  \begin{equation}
    s(t) = S_0 + \sum_{n=1}^\infty \Big(S_ne^{jn\omega_0t}+ S_{-n}e^{-jn\omega_0t}\Big).
  \end{equation}

* Working with the sign of $n$ and undoing partially the summatory, we achieve to write:

  \begin{equation}
    \begin{array}{rcl}                                                            
      s(t) & = & S_0 + \displaystyle\sum_{n=1}^\infty \Big(S_{n}e^{jn\omega_0t}+S_{-n}e^{j(-n)\omega_0t}\Big) \\                                                 
      & = & S_0 + \displaystyle\sum_{n=1}^\infty S_{n}e^{jn\omega_0t} + \displaystyle\sum_{n=-1}^{-\infty} S_ne^{jn\omega_0t}.                                   
    \end{array}                                                                   
  \end{equation}

* Finally, joining the sumatories and the outside term, we obtain the *exponential Fourier series*:

  \begin{equation}
    s(t) = \sum_{n=-\infty}^\infty S_ne^{jn\omega_0t},
    \tag{exponential_Fourier_series}
    \label{exponential_Fourier_series}
  \end{equation}

  where $S_n$ are the coefficients of the exponential Fourier series, that (using the Eqs. [Fourier_series_relation](Fourier_series_relation)) can be written as:

  <a id='Fourier_series_relation_2'></a>
  \begin{equation}                                                                
    S_n = \frac{1}{2}(a_n-jb_n).
    \tag{Fourier_series_relation_2}
    \label{Fourier_series_relation_2}
  \end{equation}

* Substituting Eq. [Fourier_series_relation_2](#Fourier_series_relation_2) in Eqs. [cosine_correlation_Lathi](#cosine_correlation_Lathi) and [sine_correlation_Lathi](#sine_correlation_Lathi), we get that:

  <a id='coeff_Fourier_exponential_series'></a>
  \begin{equation}
    \begin{array}{rl}
      S_n & = \displaystyle\frac{1}{2}\Big(\frac{2}{T}\int_{t_0}^{t_0+T} s(t)\cos(n\omega_0t)dt -j\frac{2}{T}\int_{t_0}^{t_0+T} s(t)\sin(n\omega_0t) dt \Big)\\
     & = \displaystyle\frac{1}{T}\Big(\int_{t_0}^{t_0+T} s(t)\cos(n\omega_0t)dt -j\int_{t_0}^{t_0+T} s(t)\sin(n\omega_0t)dt\Big)\\
     & = \displaystyle\frac{1}{T}\int_{t_0}^{t_0+T}s(t)e^{-jn\omega_0t} dt
    \end{array}
    \tag{coeff_Fourier_exponential_series}                                                                
    \label{coeff_Fourier_exponential_series}
  \end{equation}

## About infinite time-domain support of Fourier series

* Any (finite) signal $s(t)$ defined in the finite interval $(t_0, t_0+T)$ can be represented by any of the Fourier series (Eqs. [trigonometric_Fourier_series](#trigonometric_Fourier_series) and [exponential_Fourier_series](#exponential_Fourier_series)).

* On the other hand, the basis functions used in the series are periodic:

  $$
    e^{j\pi \omega_0t} = e^{j\pi \omega_0(t+T)},
  $$
 
  where $T=\frac{\omega_0}{2\pi}$.

* Therefore, Eqs. [trigonometric_Fourier_series](#trigonometric_Fourier_series) and [exponential_Fourier_series](#exponential_Fourier_series) are valid when $s(t)$ is periodic and infinite.

## Fourier spectrum

* The existence of a Fourier exponential series of a signal means that it can be represented by the set of complex coefficients

  $$
    \{S_0, S_{\pm 1}, S_{\pm 2}, \cdots\}
  $$

  which represent the amplitude of each angular frequency

  $$
    \{0, \pm \omega_0, \pm 2\omega_0, \cdots \},
  $$
  
  where

  $$
    \omega_0 = \frac{2\pi}{T}
  $$

  and $T$ is the period of the signal.

## Continuous time Fourier transform

* As we have seen, any periodic signal $s_T(t)$ (with period $T$) can be represented by a Fourier series. This reasoning can be extended to non periodic signals $s(t)$ if we assume that for these signals $T=\infty$. Let

  \begin{equation}
    \lim_{T\rightarrow\infty}s_T(t) = s(t).
  \end{equation}

* As a periodic signal, we can express $s_T(t)$ as:

  <a id='series_representation'></a>
  \begin{equation}
    s_T(t)=\sum_{n=-\infty}^\infty S_ne^{jn\omega_0t},
    \tag{series_representation}
    \label{series_representation}
  \end{equation}

  where

  <a id='series_coeffs'></a>
  \begin{equation}
    S_n = \frac{1}{T}\int_{t_0}^{t_0+T} s_T(t)e^{-jn\omega_0t}dt.
    \tag{series_coeffs}
    \label{series_coeffs}
  \end{equation}

* As can be seen, if $T\rightarrow\infty$, $\omega_0\rightarrow 0$ and the spectrum becomes more dense. It can be also seen in Eq. [series_coeffs](#series_coeffs) that at the same time, the amplitude of the Fourier coefficients decrease with $T$. In the limit, the Fourier spectrum exists for any $\omega$ and can be considered continuous. We will emphasize this with the change $n\omega_0 = \omega_n$ and $S_n = \frac{S(\omega_n)}{T}$ in Eqs. [series_representation](#series_representation) and [series_coeffs](#series_coeffs):

  <a id='series_representation_2'></a>
  \begin{equation}
    s_T(t)=\frac{1}{T}\sum_{n=-\infty}^\infty S(\omega_n)e^{j\omega_nt}
    \tag{series_representation_2}
    \label{series_representation_2}
  \end{equation}

  <a id='series_coeffs_2'></a>
  \begin{equation}
    S(\omega_n) =\int_{-\infty}^{\infty} s_T(t)e^{-j\omega_nt}dt.
    \tag{series_coeffs_2}
    \label{series_coeffs_2}
  \end{equation}

* Replacing $T=\frac{2\pi}{\omega_0}$ in [series_representation_2](#series_representation_2) takes us to:

  <a id='series_representation_3'></a>
  \begin{equation}
    f_T(t)=\frac{1}{2\pi}\sum_{n=-\infty}^\infty S(\omega_n)e^{j\omega_nt}\omega_0.
    \tag{series_representation_3}
    \label{series_representation_3}
  \end{equation}

* Sumarizing, if $T\rightarrow\infty$, $s_T(t)=s(t)$, $\omega_0\rightarrow 0$ and there are infinite $\omega_n$. Therefore, the sumatory in Eq. [series_representation_3](#series_representation_3) becomes a integral (disapearing the subscripts $n$ and replacing $\omega_0$ by $d\omega$), resulting that (see [Lathi](https://scholar.google.es/scholar?hl=es&as_sdt=0%2C5&q=lathi+linear+systems+and+signals&btnG=&oq=+lathi+li)):

  <a id='inverse_Fourier_transform'></a>
  \begin{equation}
    s(t)=\frac{1}{2\pi}\int_{-\infty}^\infty S(\omega)e^{j\omega t}d\omega.
    \tag{inverse_Fourier_transform}
    \label{inverse_Fourier_transform}
  \end{equation}

  <a id='Fourier_transform'></a>
  \begin{equation}
    S(\omega) = \int_{-\infty}^{\infty} s(t)e^{-j\omega t}dt.
    \tag{Fourier_transform}
    \label{Fourier_transform}
  \end{equation}

* $S(\omega)$ represents the amplitude of the infinite spectral coefficients of $s(t)$ and it known as the *spectral density function*. Eqs. [Fourier_transform](#Fourier_transform) and [inverse_Fourier_transform](#inverse_Fourier_transform) are known as the pair of Fourier transforms ([Oppenheim](https://scholar.google.es/scholar?hl=es&as_sdt=0%2C5&q=oppenheim+discrete+time+signal+processing&btnG=&oq=Oppenheim+Discrete-time+sign)). We represent this relationship between $s(t)$ and $S(\omega)$ as:

  \begin{equation}
    {\cal{F}}[s(t)] = S(\omega).
  \end{equation}

## The time- and frequency-domains

\begin{equation}
  S(\omega) = |S(\omega)|e^{j\phi(\omega)} = \mathrm{Re}(\omega) + j\mathrm{Im}(\omega),
\end{equation}

where

<a id='module'></a>
\begin{equation}
  |s(\omega)| = \sqrt{\mathrm{Re}(\omega)^2+\mathrm{Im}(\omega)^2},
  \tag{module}
  \label{module}
\end{equation}

and

<a id='phase'></a>
\begin{equation}
  \phi(\omega) = \arctan\frac{\mathrm{Im}(\omega)}{\mathrm{Re}(\omega)}.
  \tag{phase}
  \label{phase}
\end{equation}

## Fourier transform of the Rectangular Signal $Ag_\tau(t)$: the Sinc signal

* Let

  <a id='funcion_cuadrada'></a>
  <img src="data/cuadrada.png" width=400>

* The Fourier transform can be found as.

  $$
    \begin{array}{ll}
      \cal{F}\big(Ag_\tau(t)\big) & = \displaystyle\int_{-\frac{\tau}{2}}^{\frac{\tau}{2}} Ae^{-j\omega t}dt \\\\
      & = \displaystyle\frac{A}{j\omega}e^{-j\omega t}\displaystyle|_{-\frac{\tau}{2}}^{\frac{\tau}{2}\
}\\\\
      & =
      \displaystyle\frac{A}{j\omega}(e^{j\omega\frac{\tau}{2}}-e^{-j\omega\frac{\tau}{2}})\\\\
      & \textrm{(considering that } e^{jx}-e^{-jx}=2j\sin(x)) \\\\
      & = A\tau\displaystyle\frac{\sin(\omega\frac{\tau}{2})}{\omega\frac{\tau}{2}}\\\\
      & \textrm{considering that } \text{Sinc}(x)=\frac{\sin(x)}{x} \text{is the SINne Cardinal function} \\\\
      & = A\tau\textrm{Sinc}(\omega\frac{\tau}{2}).
    \end{array}
  $$

* Therefore:

  <a id='sinc'></a>
  \begin{equation}
    \cal{F}\big(Ag_\tau(t)\big)=A\tau\textrm{Sinc}(\omega\frac{\tau}{2})
    \label{sinc}
    \tag{sinc}
  \end{equation}

  <img src="data/funcion_muestreo.png" width=600>

* Notice that, if $\tau\rightarrow \infty$ then the spectrum of a rectangular function becomes an impulse function at $\omega=0$, and if $\tau\rightarrow 0$, the spectrum is a constant function.

* The function
  
  <a id='pure_sinc'></a>
  \begin{equation}
    \mathrm{Sinc}(x) = \left\{
      \begin{array}{ll}
        \frac{\sin(x)}{x} & \mbox{if $x \neq 0$};\\
        1 & \mbox{if $x = 0$}.
        \end{array} \right.
     \label{pure_sinc}
     \tag{pure_sinc}
  \end{equation}

 is also called *The Sampling Function*.

## Dirac delta signal

* Also known as the [unit impulse signal](https://en.wikipedia.org/wiki/Dirac_delta_function).

### Definitions

1. Dirac Delta Signal $\delta(t)$ is an impulse of energy $1$, infinitely narrow and high, whose area is $1$:

   <a id='delta_area'></a>
   \begin{equation}
     \left\{
       \begin{array}{ll}
         \displaystyle\int_{-\infty}^\infty\delta(t)dt=1 & \text{if $t=0$}\\
         0 & \text{otherwise.}
       \end{array}
     \right.
     \tag{delta_area}
     \label{delta_area}
   \end{equation}

   <img src="data/delta.png" width=350>

2. The area of $\delta(t)$ multiplied by a signal $s(t)$ is the value of $s(t)$ where $\delta(t)$ is defined:

   <a id='product_signal_delta_area'></a>
   \begin{equation}
     \int_{-\infty}^\infty\delta(t)s(t)dt =
     s(0)\int_{-\infty}^\infty\delta(t)dt = s(0),
     \tag{product_signal_delta_area}
     \label{product_signal_delta_area}
   \end{equation}

   <a id='product_signal_shifted_delta'></a>
   \begin{equation}
     \int_{-\infty}^\infty\delta(t-t_0)s(t)dt = s(t_0).
     \tag{product_signal_shifted_delta_area}
     \label{product_signal_delta_shifted_area}  
   \end{equation}

### Generation

* The function $\delta(t)$ represents a signal physically impossible to generate, that only can be get in the limit of other functions:

  <a id='delta_sinc'></a>
  \begin{equation}
    \delta(t) =
    \lim_{\tau\rightarrow\infty}\frac{\tau}{\pi}\textrm{Sinc}(\tau t).
    \tag{delta_sinc}
    \label{delta_sinc}
  \end{equation}

  <img src="data/funcion_muestreo.png" width=500>

## Fourier transform of the constant signal $1$

<a id='Fourier_transform_constant_signal'></a>
\begin{equation}
  {\cal F}[1] = 2\pi\delta(\omega).
  \tag{Fourier_transform_constant_signal}                                                                                                                      
  \label{Fourier_transform_constant_signal} 
\end{equation}

<img src="data/1_delta.png" width=800>

### Proof

* By definition (see Eq. [rectangular signal](#funcion_cuadrada)):

  \begin{equation}
    \displaystyle\lim_{\tau\rightarrow\infty} g_\tau(t) = 1.
  \end{equation}

* Applying the Fourier transform to both sides:

  \begin{equation}
    {\cal F}[1] = {\cal F}[\displaystyle\lim_{\tau\rightarrow\infty} g_\tau(t)].
  \end{equation}

* The Fourier transform of a limit is the limit of the Fourier Transform:

  \begin{equation}
    {\cal F}[\displaystyle\lim_{\tau\rightarrow\infty} g_\tau(t)] = \displaystyle\lim_{\tau\rightarrow\infty}{\cal F}[g_\tau(t)]
  \end{equation}

* The Fourier transform of a rectangular signal is a Sinc (see Eq. [sinc](#sinc)):

  \begin{equation}
    \displaystyle\lim_{\tau\rightarrow\infty}{\cal F}[g_\tau(t)] = \displaystyle\lim_{\tau\rightarrow\infty}\tau\text{Sinc}(\frac{\tau}{2}\omega)
  \end{equation}

* A Sinc with period $\infty$ becomes a Delta (see Eq. [delta_sinc](#delta_sinc)):

  \begin{equation}
   \displaystyle\lim_{\tau\rightarrow\infty}\tau\text{Sinc}(\frac{\tau}{2}\omega) = \displaystyle2\pi\lim_{\tau\rightarrow\infty} \frac{\tau}{2\pi}\textrm{Sinc}(\frac{\tau}{2}\omega) = 2\pi\delta(\omega).
  \end{equation}

## Fourier transform of the complex exponential function $e^{j\omega_0t}$

(Note, in the nature, there not exist complex signals).

<a id='Fourier_transform_complex_exponential'></a>
\begin{equation}
  {\cal F}[e^{jw_0t}] = 2\pi\delta(\omega-\omega_0).
  \tag{Fourier_transform_complex_exponential}                                                                                                                      
  \label{Fourier_transform_complex_exponential}
\end{equation}

<img src="data/TFexpo.png" width=400>

### Proof:

* By definition of Fourier transform of signal $s(t)$ (see Eq. [Fourier_transform](#Fourier_transform)):

  \begin{equation}
    S(\omega_0) = \int_{-\infty}^\infty s(t)e^{-j\omega_0t}dt,
  \end{equation}

  and by definition of inverse Fourier transform of $S(\omega)$ (see Eq. [inverse_Fourier_transform](#inverse_Fourier_transform)):

  \begin{equation}
    s(t) = \frac{1}{2\pi} \int_{-\infty}^\infty S(\omega)e^{j\omega t}d\omega.
  \end{equation}

* Substituting the second equation in the first one, we get that:

  \begin{equation}
    S(\omega_0) = \int_{-\infty}^\infty \big( \frac{1}{2\pi} \int_{-\infty}^\infty S(\omega)e^{j\omega t}d\omega \big) e^{-j\omega_0t}dt = \int_{-\infty}^\infty S(\omega) \big( \frac{1}{2\pi} \int_{-\infty}^\infty e^{j\omega t}e^{-j\omega_0 t}dt\big) d\omega.
  \end{equation}

* On the other hand, we know (see Eq. [product_signal_delta_shifted_area](#product_signal_delta_shifted_area)) that:

  \begin{equation}
    S(\omega_0) = \int_{-\infty}^\infty S(\omega)\delta(\omega-\omega_0)d\omega.
  \end{equation}

* Therefore, comparing both expressions, we get that:

  \begin{equation}
    \frac{1}{2\pi} \int_{-\infty}^\infty e^{j\omega t}e^{-j\omega_0t}dt = \delta(\omega-\omega_0),
 \end{equation}

  or that:

  \begin{equation}
    \int_{-\infty}^\infty e^{j\omega t}e^{-\omega_0 t}dt = 2\pi\delta(\omega-\omega_0),
  \end{equation}

* The first part of this equation is the Fourier Transform of ${\cal F}(e^{j\omega_0 t})$. Therefore:

  \begin{equation*}
    {\cal F}[e^{j\omega_0t}] = 2\pi\delta(\omega-\omega_0).
  \end{equation*}


## Fourier transform of the sine signal

<a id='Fourier_transform_sine'></a>
\begin{equation}
  {\cal F}[\sin(\omega_0t)] = j\pi\big(\delta(\omega+\omega_0)-\delta(\omega-\omega_0)\big).
  \tag{Fourier_transform_sine}
  \label{Fourier_transform_size}
\end{equation}

<img src="data/TFseno.png" width=800>

### Proof

* As can be seen in [List of trigonometric identities](https://en.wikipedia.org/wiki/List_of_trigonometric_identities#Relation_to_the_complex_exponential_function):

  \begin{equation}                                                                                                                   
    \sin(\omega_0t) = \frac{\displaystyle e^{j\omega_0t}-e^{-j\omega_0t}}{\displaystyle 2j}.                                                         
  \end{equation}

* Using the [linearity property](http://fourier.eng.hmc.edu/e101/lectures/handout3/node2.html) of the Fourier transform:

  \begin{equation}
    {\cal F}[\sin(\omega_0t)] = \displaystyle\frac{1}{2j}({\cal F}[e^{j\omega_0t}]-{\cal F}[e^{-j\omega_0t}])
 \end{equation}

* Applying Eq. [Fourier_transform_complex_exponential](#Fourier_transform_complex_exponential) to this expression we get that:

\begin{equation*}
  {\cal F}[\sin(\omega_0t)] = \displaystyle\frac{1}{2j}\big(2\pi\delta(\omega-\omega_0)-2\pi\delta(\omega+\omega_0)\big) = j\pi\big(\delta(\omega+\omega_0)-\delta(\omega-\omega_0)\big).
\end{equation*}

## Fourier transform of the cosine signal

<a id='Fourier_transform_cosine'></a>
\begin{equation}
\begin{array}{rl}
  {\cal F}[\cos(\omega_0t)] = j\pi\big(\delta(\omega+\omega_0)+\delta(\omega-\omega_0)\big).
\end{array}
\tag{Fourier_transform_cosine}                                                                                                                     
\label{Fourier_transform_cosine}
\end{equation}

<img src="data/TFcoseno.png" width=800>

### Proof

\begin{equation}                                                                                                                   
  \cos(w_0t) = \frac{\displaystyle e^{j\omega_0t}+e^{-j\omega_0t}}{\displaystyle 2j}.                                                         
\end{equation}

\begin{equation}
\begin{array}{rl}
  {\cal F}[\cos(\omega_0t)] & = \displaystyle\frac{1}{2j}({\cal F}[e^{j\omega_0t}]+{\cal F}[e^{-j\omega_0t}])\\\\
  & = \displaystyle\frac{1}{2j}\big(2\pi\delta(\omega-\omega_0)+2\pi\delta(\omega+\omega_0)\big)\\\\
  & = j\pi\big(\delta(\omega+\omega_0)+\delta(\omega-\omega_0)\big).
\end{array}
\end{equation}


## Fourier transform of a periodic signal

* Any periodic signal can be expressed in its exponential Fourier series (see Eq. [exponential_Fourier_series](#exponential_Fourier_series)):

  \begin{equation}
    \begin{array}{rl}                                                                                                               
      {\cal F}[s(t)] & = {\cal F}[\displaystyle\sum_{n=-\infty}^\infty S_n e^{jn\omega_ot}]\\\\
      & = \displaystyle\sum_{n=-\infty}^\infty S_n{\cal F}[e^{jn\omega_ot}].                                                             
    \end{array}
  \end{equation}

* Applying Eq. [Fourier_transform_complex_exponential](#Fourier_transform_complex_exponential):

\begin{equation}                                                                                                                  
  {\cal F}[f(t)] = 2\pi\sum_{n=-\infty}^\infty S_n\delta(\omega-n\omega_0).                                                                 
  \tag{FTps}                                                                                                                      
  \label{eq:TFps}                                                                                                                 
\end{equation}

* This relation is very important because it establishes that the spectral density function (the Fourier transform) of a periodic signal is composed of pulses located in the harmonic frequencies (multiplying frequencies of the fundamental frequency $\omega_0$) of the signal, being the energy of each impulse $2\pi$ multiplied by the value of the corresponding coefficient of the exponential Fourier series.

## Fourier Transform of train of equidistant unit impulses

<a id='${\cal F}[\delta_T(t)]$'></a>
\begin{equation}
  {\cal F}[\delta_T(t)] = \omega_0\delta_{\omega_0}(\omega).
  \tag{${\cal F}[\delta_T(t)]$}
  \label{${\cal F}[\delta_T(t)]$}
\end{equation}

<img src="data/TFtren_de_impulsos.png" width=800>

### Proof

* Let:

  $$
    \delta_T(t) = \sum_{n=-\infty}^{\infty}\delta(t-nT).                        
    \tag{$\delta_T$}                                                            
    \label{eq:delta_T}
  $$

* As any other signal (see Eq. [exponential_Fourier_series](#exponential_Fourier_series)), $\delta_T(t)$ can be written as:

  \begin{equation}
    \delta_T(t)=\sum_{n=-\infty}^{\infty}F_ne^{jn\omega_0t}     
  \end{equation}

  where

  \begin{equation}
    \omega_0 = \frac{2\pi}{T}
  \end{equation}

  and

  \begin{equation}
    F_n = \frac{1}{T}\int_{-\frac{T}{2}}^{\frac{T}{2}}\delta_T(t)e^{-jn\omega_0t}dt.
  \end{equation}

* The signal $\delta_T(t)$ in the interval $[-\frac{T}{2},\frac{T}{2}]$ is equal to $\delta(t)$. Therefore:

  $$
    F_n = \frac{1}{T}\int_{-\frac{T}{2}}^{\frac{T}{2}}\delta(t)e^{-jn\omega_0t}dt.
  $$

* Considering that the $\delta(t)$ have energy only in $t=0$:

  $$
  \frac{1}{T}\int_{-\frac{T}{2}}^{\frac{T}{2}}\delta(t)e^{-jnw_0t}dt =
    \frac{1}{T}\int_{-\infty}^{\infty}\delta(t)e^{-jnw_0t}dt.
  $$

* Applying that (see Eq. [product_signal_delta_area](#product_signal_delta_area)):

  $$
  \int_{-\infty}^\infty \delta(t)e^{-jn\omega_0 t}dt = e^{-jn\omega_0 0}\int_{-\infty}^\infty \delta(t)dt = 1,
  $$

  we get that:

  $$
    F_n = \frac{1}{T},
  $$

  and therefore, that

  $$
  \delta_T(t) = \sum_{n=-\infty}^{\infty} \frac{1}{T} e^{jn\omega_0t} = \frac{1}{T}\sum_{n=-\infty}^{\infty}e^{jn\omega_0t}.
  $$

* Considering that (see Eq. [Fourier_transform_complex_exponential](#Fourier_transform_complex_exponential)):

  \begin{equation*}
   {\cal F}[e^{jn\omega_0 t}] = 2\pi\delta(\omega-n\omega_0),
  \end{equation*}

  we get that:

  $$
    \begin{array}{rl}
      {\cal F}[\delta_T(t)] & =
      2\pi\displaystyle\sum_{n=-\infty}^\infty\frac{1}{T}\delta(\omega-n\omega_0)\\\\
      & =
      \displaystyle\frac{2\pi}{T}\sum_{n=-\infty}^\infty\delta(\omega-n\omega_0)\\\\
      & = \omega_0\displaystyle\sum_{n=-\infty}^\infty\delta(\omega-n\omega_0)\\\\
      & = \omega_0\delta_{\omega_0}(\omega).
    \end{array}
  $$

## Fourier transform of a time-shifted signal $s(t-t_0)$

<a id='Fourier_transform_time-shifted_signal'></a>
\begin{equation}
  {\cal F}[s(t-t_0)] = S(\omega)e^{-j\omega t_0}.
  \tag{Fourier_transform_time-shifted_signal}
  \label{Fourier_transform_time-shifted_signal}
\end{equation}

### Proof

* Using Eq. [Fourier_transform](#Fourier_transform),

  $$
  {\cal F}[s(t-t_0)] = \int_{-\infty}^\infty s(t-t_0)e^{-j\omega t}dt.
  $$

* Let $x=t-t_0$. Then:

  \begin{equation*}
    \begin{array}{rl}                                                               
      {\cal F}[s(t-t_0)] & = \displaystyle\int_{-\infty}^\infty s(x)e^{-j\omega(x+t_0)}dx\\\\                                                      
      & = \displaystyle\int_{-\infty}^\infty s(x)e^{-j\omega x}dx\cdot e^{-j\omega t_0}\\\\
      & = S(\omega)e^{-j\omega t_0}.
    \end{array}
  \end{equation*}

## Inverse Fourier transform of a frequency-shifted signal $S(\omega-\omega_0)$

<a id='inverse_Fourier_transform_-frequency-shifted_signal'></a>
\begin{equation}
  S(\omega-\omega_0) = {\cal F}[s(t)e^{j\omega_0 t}]
  \tag{inverse_Fourier_transform_frequency-shifted_signal}
  \label{inverse_Fourier_transform_frequency-shifted_signal}
\end{equation}

### Proof

* Using Eq. [inverse_Fourier_transform](#inverse_Fourier_transform),

  \begin{equation*}
    S(\omega-\omega_0) = \displaystyle\int_{-\infty}^\infty s(t) e^{j(\omega-\omega_0)t} dt =\\ \displaystyle\int_{-\infty}^\infty \Big(s(t) e^{j\omega_0t}\Big) e^{-j\omega t} dt =\\ {\cal F}[s(t)e^{j\omega_0 t}]
  \end{equation*}

## Signals convolution: definition

<a id='signals_convolution'></a>
\begin{equation}                                                              
  s_1(t)*s_2(t)=\int_{-\infty}^\infty s_1(\tau)s_2(t-\tau)d\tau.              
  \tag{signals_convolution}                                      
  \label{signals_convolution}                                                      
\end{equation}

## Time-domain convolution = Frequency-domain multiplication

\begin{equation}
  s_1(t)*s_2(t) = {\cal F}^{-1}[S_1(\omega)S_2(\omega)].
  \tag{time-domain_convolution}
  \label{time-domain_convolution}
\end{equation}

### Proof:

\begin{equation}
  \begin{array}{rl}
    {\cal F}[s_1(t)*s_2(t)] & = \displaystyle\int_{-\infty}^\infty \Big[\int_{-\infty}^\infty s_1(\tau)s_2(t-\tau)d\tau\Big]e^{-j\omega t}dt\\\\
  & ~\text{(using Fourier transform of a time-shifted signal)} \\\\
  & = \displaystyle\int_{-\infty}^\infty s_1(\tau)\Big[\underbrace{\int_{-\infty}^\infty s_2(t-\tau)e^{-j\omega t}dt}_{S_2(\omega)e^{-j\omega\tau} }\Big]d\tau \\\\
& = \displaystyle\int_{-\infty}^\infty
s_1(\tau)S_2(\omega)e^{-j\omega\tau}d\tau \\\\
& = S_2(\omega)\displaystyle\int_{-\infty}^\infty s_1(\tau)e^{-j\omega\tau}d\tau\\\\
& = S_1(\omega)S_2(\omega).
\end{array}
\end{equation}

## Frequency-domain convolution = Time-domain multiplication

\begin{equation}
  s_1(t)s_2(t) = {\cal F}^{-1}\Big[\frac{1}{2\pi}\big(S_1(\omega)*S_2(\omega)\big)\Big].
  \tag{frequency-domain_convolution}
  \label{frequency-domain_convolution}
\end{equation}

### Proof

\begin{equation}
  \begin{array}{rl}
    {\cal F}^{-1}\Big[\displaystyle\frac{1}{2\pi}\big(S_1(\omega)*S_2(\omega)\big)\Big] & =
  \displaystyle\frac{1}{2\pi}\displaystyle\int_{-\infty}^\infty
  \Big[\frac{1}{2\pi}\big(S_1(\omega)*S_2(\omega)\big)\Big]e^{j\omega t}d\omega \\\\
  & \text{(using the definition of convolution)} \\\\
  & = \displaystyle\frac{1}{2\pi}\displaystyle\int_{-\infty}^\infty
  \Big[\frac{1}{2\pi}\big(\displaystyle\int_{-\infty}^\infty S_1(\tau)S_2(\omega-\tau)d\tau\big)\Big]e^{j\omega t}d\omega \\\\
  & = \displaystyle\frac{1}{2\pi}\displaystyle\int_{-\infty}^\infty S_1(\tau)\underbrace{\Big[\frac{1}{2\pi}\displaystyle\int_{-\infty}^\infty S_2(\omega-\tau)e^{j\omega t}d\omega\Big]}_{s_2(t)e^{j\tau t}} d\tau \\\\
  & \text{(using inverse Fourier transform of a frequency-shifted signal)} \\\\
  & = \displaystyle\frac{1}{2\pi}\displaystyle\int_{-\infty}^\infty S_1(\tau)s_2(t)e^{j\tau t}d\tau \\\\
  & = f_2(t)\Big[\displaystyle\frac{1}{2\pi}\displaystyle\int_{-\infty}^\infty
  F_1(\tau)e^{j\tau t}d\tau\Big] \\\\
  & = f_2(t)f_1(t).
\end{array}
\end{equation}

## Covolution of a signal with a delta

\begin{equation}
  s(t)*\delta(t) = s(t).
  \tag{convolution_signal_delta}
  \label{convolution_signal_delta}
\end{equation}

### Proof

\begin{equation}
  s(t)*\delta(t)={\cal F}^{-1}[S(\omega)\Delta(\omega)] = {\cal F}^{-1}[S(\omega)\cdot 1] = s(s).
\end{equation}

## Convolution of a signal with a time-shifted delta

\begin{equation}
  s(t)*\delta(t-t_0) = s(t-t_0).
  \tag{convolution_signal_shifted-delta}
  \label{convolution_signal_shifted-delta}
\end{equation}

### Proof

\begin{equation}
  \begin{array}{rcl}
  f(t)*\delta(t-t_0) & = &
  {\cal F}^{-1}\big[F(w)(\Delta(w)e^{-jwt_0})\big]\\\\
  & = & {\cal F}^{-1}\big[(F(w)e^{-jwt_0})\Delta(w)\big]\\\\
  & = & {\cal F}^{-1}\big[{\cal F}[f(t-t_0)]\underbrace{\Delta(w)}_{1}\big]\\\\
  & = & {\cal F}^{-1}\big[{\cal F}[f(t-t_0)]\big]\\\\
  & = & f(t-t_0).
  \end{array}
\end{equation}
